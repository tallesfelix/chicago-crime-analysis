```python
import geopandas as gpd
import emoji
import pandas as pd
from ipyleaflet import *
import osmnx as ox
import os
import matplotlib.pyplot as plt
import networkx as nx
import seaborn as sns
from os import listdir
gpd.io.file.fiona.drvsupport.supported_drivers['LIBKML'] = 'rw'
%matplotlib inline
```

```python
path = './crime_data/city-of-london-stop-and-search/'
folders = listdir(path)

data_list = []

for folder in folders:
    data_path = os.path.join(path,folder)
    file = listdir(data_path)[0]
    data_list.append(pd.read_csv(os.path.join(data_path,file)))
    
sas = pd.concat(data_list, axis=0, ignore_index=True)
```

```python
def create_day_of_week_column(dataset, datetime_column='datetime'):
    '''Creates a column with the day of week from datetime'''
    dataset['week day'] = dataset[datetime_column].apply(lambda datetime: datetime.dayofweek)
    
    day_map = {0:'Seg',1:'Ter',2:'Qua',3:'Qui',4:'Sex',5:'Sab',6:'Dom'}
    dataset['week day'] = dataset['week day'].map(day_map)
    
    return dataset
```

```python
sas = create_day_of_week_column(sas, datetime_column="Date")
```

```python
def create_geopandas(pd_dataframe, lat=None, lon=None):
    geodata = gpd.GeoDataFrame(pd_dataframe, 
                               geometry=gpd.points_from_xy(pd_dataframe[lon], pd_dataframe[lat]),
                                crs='epsg:4326')
    
    return geodata
```

```python
geo_sas = create_geopandas(sas, lat='Latitude', lon='Longitude')
```

```python
area = gpd.read_file('./crime_data/city-of-london.kml')['geometry'].values[0]
```

```python
graph = ox.graph_from_polygon(area, network_type='drive')
```

```python
nodes, edges = ox.graph_to_gdfs(graph)
```
